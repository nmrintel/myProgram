```python
import csv
import numpy as np
import math

def trancate(number,deximals=0):
    
    multiplier = 10 ** deximals
    return int(number * multiplier) / multiplier

def makeCluster():
    with open('data1.csv', 'w',newline="") as f:
        writer = csv.writer(f)
        writer.writerow(['height', 'weight'])
        numCluter = 4
        for i in range(numCluter):
            xc = np.random.random()
            yc = np.random.random()
            for i in range(int(100/numCluter)):
                x = xc + 0.5* (np.random.rand() - 0.5)
                y = yc + 0.5 * (np.random.rand() - 0.5)
                x = trancate(x, 4)
                y = trancate(y, 4)
        writer.writerow([x, y])

def quadraticData():
    with open('data2.csv','w',newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["x","y"])
        x = np.random.rand(100,1) -0.5
        y = 4* x**2 + x +0.5*np.random.rand(100,1)
        writer.writerows(np.c_[x,y])
```
