```python
import numpy as np
from skimage import data,io
from scipy import ndimage as ndi
```

```python
coins = data.coins()
io.imshow(coins)
```

```python
img = np.zeros(coins.shape)
mask = coins[:,:] > 100
img[mask] = 255
io.imshow(img)
```

```python
fill_coins = ndi.binary_fill_holes(img)
io.imshow(fill_coins)
```

```python
square_img = np.random.rand(500,500)
io.imshow(square_img)
```

```python
shape = np.array(square_img.shape)
center = shape/2
radius = shape[0]/2

mask = np.zeros(shape)
for i in range(shape[0]):
    for j in range(shape[1]):
        if ((i-center[0])**2 + (j-center[1])**2 > radius**2):
            square_img[i][j] = 0

io.imshow(square_img)
```
