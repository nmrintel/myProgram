```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("titanic/train.csv")
table = table[['Survived','Age']]
table = np.array(table)
dif = []
```

```python
def child(age):
    table2 = table.copy()
    table2[:,1] = table2[:,1] > age

    adult_alive = 0
    child_alive = 0
    adult_dead = 0
    child_dead = 0

    for i in range(len(table2)):
        if(table2[i][1] == 0):
            #child
            if(table2[i][0] == 0):
                child_dead+=1
            else:
                child_alive+=1
        else:
            #adult
            if(table2[i][0] == 0):
                adult_dead+=1
            else:
                adult_alive+=1
    
    ratio_child = child_alive/(child_alive+child_dead)
    ratio_adult = adult_alive/(adult_alive+adult_dead)
    dif.append([ratio_child-ratio_adult])
```

```python
for i in range(5,60):
    child(i)
    
plt.plot(dif)
```
